### Run Function

Next, let us set up starting date and ending date for the data that you like and specify a stock ticker you want to look at.

Note: one can specify "verbose=True" which will print out a brief summary:
- stock price charts with adjusted closing price and moving averages
- time-series plot of signals (positive means sell and negative means buy assuming using reversal strategy)
- expected return
- expected risk (measured by volatility)
- tail of buy signal (developed by myself; and transformation of moving averages; for more, please go to my research site [here](https://yinscapital.com/research/) and click on "Buy Signal from Limit Theorem".)

```python
# Input
ticker = 'BA'
start_date = '2015-01-01'
end_date   = '2020-01-18'

# Output (function creates a dictionary):
# Run function and save verything in a *temp* variable (which will become a dictonary)
temp = YinsMM.YinsTimer(
    start_date, end_date, ticker, figsize=(8,4), LB=-0.09, UB=0.17,
    plotGraph=True, verbose=True, printManual=False, gotoSEC=True)
```

Once the code finishes running, we can start to print the result by looking at basic statistics such as expected return and risk. This will be exactly the same numbers from printed results from above.

```python
print(f'Ticker: {ticker}')
print(f"Expted Return: {round(temp['estimatedReturn'], 4)}")
print(f"Expted Risk (Volatility): {round(temp['estimatedRisk'], 4)}")
print(f"Reward-Risk Ratio (Daily Data): {round(temp['estimatedReturn'] / temp['estimatedRisk'], 4)}")
```
